```python
rice_state=df.groupby('State Name')['RICE PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).reset_index()
rice_state
```

```python
#Bar plot
fig = plt.figure(figsize = (8, 5))
plt.bar(df['State Name'],df['RICE PRODUCTION (1000 tons)'], color ='red',width =0.4)
plt.xlabel("State Name")
plt.ylabel("RICE PRODUCTION (1000 tons)")
plt.title("South_India Rice Productions By States")
plt.show()
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
fig = px.pie(rice_state, names='State Name', values='RICE PRODUCTION (1000 tons)',
             title='Total Rice Production State_wise', hole=0.0,height=600,width=800)
fig.show()
```

```python
Groundnut_state=df.groupby('State Name')['GROUNDNUT PRODUCTION (1000 tons)'].sum().sort_values(ascending=False).reset_index()
Groundnut_state
```
